# Lesson 0: Convergence of Errors
-----------------------  
Existing numerical methods for solving a linear system of equations, $Au=f$, fall into two large categories: *direct methods* and *iterative (or relaxation) methods*. Iterative methods show an interesting property for the convergence of errors: the convergence rates depend on the frequencies of the errors. This interesting property is the basis of multigrid methods.  We will start this multigrid tutorial with an introduction of iterative methods and a discussion of error.

We first review classical iterative methods. In the second section, we define the example problem we will use to examine the convergence of errors. Finally, we discuss the convergence of errors in the third section.

## 0-1 Basic Iterative Methods
Iterative methods begin with an initial guess at a solution and try to improve the approximated solution through iterations. Ideally, the approximated solutions generated from each iteration should converge to the exact solution. 

We review two methods here -- the weighted Jacobi method and the red-black Gauss-Seidel method.

### 0-1-1 Weighted Jacobi Method
If $\mathbf{v}^{(m)}$ represents the approximated solutions at $m_{th}$ iteration, $\mathbf{v}^{(m+1)}$ in Jacobi Method is:

$$
v{_i^{(m+1)}} = [b_i - (A_{i,0}v{_0^{(m)}}+\cdots+A_{i,i-1}v{_{i-1}^{(m)}}+A_{i,i+1}v{_{i+1}^{(m)}}+\cdots+A_{i,n}v{_n^{(m)}})] / A_{i,i}
$$

Or we can express it in matrix form:

$$
\mathbf{v}^{(m+1)}=D^{-1}(\mathbf{b} - (L+U) \mathbf{v}^{(m)})
$$

where $D$, $L$, and $U$ represent the diagonal, lower triangular, and upper triangular parts of $A$ respectively, i.e. $A = D + L + U$. 

The above matrix equation can also be written as:

$$
\mathbf{v}^{(m+1)}=\mathbf{v}^{(m)} + D^{-1}(\mathbf{b} - A\mathbf{v}^{(m)})
$$

From this matrix equation, we can see that the new approximated solution $\mathbf{v}^{(m+1)}$ is actually the old approximated solution $\mathbf{v}^{(m)}$ plus a correction term ($D^{-1}(\mathbf{b} - A\mathbf{v}^{(m)})$). The weighted Jacobi method tries to adjust the amount of correction by defining a parameter $\omega$ such that:

$$
\mathbf{v}^{(m+1)}=\mathbf{v}^{(m)} + \omega D^{-1}(\mathbf{b} - A\mathbf{v}^{(m)})
$$  
If $\omega=1$, the weighted Jacobi method is equivalent to the traditional Jacobi method. 

The Python code for weighted Jacobi method is:

```python
def WJacobi(N, A, v0, b, omg=2./3., Niter=1):
    '''
    Inputs:
        N: the dimension of A, v0, and b (integer)
        A: coefficient matrix (N by N array)
        v0: initial guess (N by 1 array)
        b: constant matrix (N by 1 array)
        omg: relaxation parameter (float)
        Niter: the number of iterations to perform
    Return:
        The approximated solution v after Niter iterations
        (N by 1 array)
    '''
    v = np.empty(N)
    DA = np.diag(A)
    for ITER in range(Niter):
        v = v0 + omg * (b - np.dot(A, v0)) / DA
        v0 = v.copy()
    return v
```

### 0-1-2 Red-Black Gauss-Seidel Method
In the weighted Jacobi method, each element in the new approximated solution $\mathbf{v}^{(m+1)}$ is constructed from elements in previous approximated solution $\mathbf{v}^{(m)}$ only. What if, instead, we use elements from the new approximated solution to calculate other elements that we are updating? That is

$$
v{_i^{(m+1)}} = [b_i - (A_{i,0}v{_0^{(m+1)}}+\cdots+A_{i,i-1}v{_{i-1}^{(m+1)}}+A_{i,i+1}v{_{i+1}^{(m)}}+\cdots+A_{i,n}v{_n^{(m)}})] / A_{i,i}
$$

This is called the Gauss-Seidel method. It can converge faster than the weighted Jacobi method, but the computation time required may still exceed that of the weighted Jacobi method as the Gauss-Seidel method is hard to parallelize.

In fact, though the equation shown above updates elements consecutively from the first to last, the order can be changed. One of the modified Gauss-Seidel methods is called the red-black Gauss-Seidel method: it updates the even elements first and then updates the odd elements, as shown below.

$$
\text{step }1:~~ v{_{2i}^{(m+1)}} = [b_{2i} - (\cdots+A_{2i,2i-2}v{_{2i-2}^{(m)}}+A_{2i,2i-1}v{_{2i-1}^{(m)}}+A_{2i,2i+1}v{_{2i+1}^{(m)}}+A_{2i,2i+2}v{_{2i+2}^{(m)}}+\cdots)] / A_{2i,2i}
$$

$$
\text{step }2:~~ v{_{2i+1}^{(m+1)}} = [b_{2i+1} - (\cdots+A_{2i,2i-2}v{_{2i-1}^{(m)}}+A_{2i,2i}v{_{2i}^{(m+1)}}+A_{2i,2i+2}v{_{2i+2}^{(m+1)}}+A_{2i,2i+3}v{_{2i+3}^{(m)}}+\cdots)] / A_{2i+1,2i+1}
$$

It only uses information from the previous approximation when updating the even elements in the first step and then uses updated information from the first step when updating the odd elements. Both step $1$ and step $2$ can be parallelized individually.

The code for the red-black Gauss-Seidel method is:

```python
def RBGaussSeidel(N, A, v0, b, Niter=1):
    '''
    Inputs:
        N: the dimension of A, v0, and b (integer)
        A: coefficient matrix (N by N array)
        v0: initial guess (N by 1 array)
        b: constant matrix (N by 1 array)
        Niter: the number of iterations to perform
    Return:
        The approximated solution v after Niter iterations
        (N by 1 array)
    '''
    DA = np.diag(A)
    for ITER in range(Niter):
        v0[0::2] = v0[0::2] + \
            (b[0::2] - np.dot(A[0::2, :], v0)) / DA[0::2]
        v0[1::2] = v0[1::2] + \
            (b[1::2] - np.dot(A[1::2, :], v0)) / DA[1::2]
    return v0
```

## 0-2 Example Problems
Before we start to discuss the convergence of errors of iterative methods, we introduce the example problem which will be used later. Consider the following ODE:

$$-u''(x)+\sigma u(x)=f(x),~~0<x<1,~~\sigma\ge0$$  
$$u(0)=u(1)=0$$

We discretize the domain into $N_s$ segments, so the number of grid points, $N_p$, is $N_s+1$, including end points. Applying central difference on the above ODE, we can obtain a linear system (including the boundary conditions): $A\mathbf{u}=\mathbf{b}$, where the vector $\mathbf{u}=(u_0,...,u_n)$ represents the exact solution on each point, and 

$$
A=
\begin{bmatrix}
1 & 0 & 0 & \cdots & \cdots & 0 \\
-1 & 2+\sigma \Delta x^2 & -1 & \ddots & &\vdots \\
\vdots & \ddots & \ddots & \ddots & \ddots & \vdots \\
\vdots & \ddots & \ddots & \ddots & \ddots & \vdots\\
\vdots & & \ddots & -1 & 2+\sigma \Delta x^2 & -1\\
0 & \cdots & \cdots & 0 & 0 & 1
\end{bmatrix},~~
\mathbf{b}=
\Delta x^2
\begin{bmatrix}
0 \\ f_1 \\ \vdots \\ \vdots \\ f_{n-1} \\ 0
\end{bmatrix}
$$

For convenience, we will assume $\sigma=0$ and $f(x)=0$ (i.e. 1D Laplace equation) later in this notebook. The exact solution of this Laplace equation is $u(x)=0$ (i.e. $\mathbf{u}=(0,...,0)$). The vector $\mathbf{v}^{(m)}$ still represents the approximated solution at $m_{th}$ iteration in iterative methods.

## 0-3 The Convergence of Errors of Iterative Methods
For a given approximated solution $\mathbf{v}^{(m)}$, the error vector $\mathbf{e}^{(m)}$ is simply $\mathbf{e}^{(m)}=\mathbf{v}^{(m)}$ in the 1D Laplace equation. The error can be presented by Fourier series, which decomposes the error into a combination of finite number of waves with different wavenumbers or frequencies. (Since the space is discrete in numerical world, the wavenumbers are discrete. And the number of the waves in the Fourier series of error is thus finite.) When error converges to zero, it means all the waves also converge to zero. If the convergence rates of the waves are not a constant and depends on the wavenumbers, then the convergence rate of the error depends on the waves that converge most slowly. 

In order to find out the relationships between the convergence rates and the wavenumbers, we make the initial guess, $v_{initial}$, to be a wave with a single wavenumber. As mentioned in the last paragraph, the approximated solution is right the error in 1D Laplace equation. This means now our initial error, $e_{initial}$ ($=v_{initial}$), only contains a single wavenumber. Let's see how fast the iterative methods eliminate errors with different wavenumbers. 

First, for convenience, we write the initializations into functions.

```python
def Init1DLaplaceAb(N):
    '''
    Initialize matrix A and vector b
    for the 1D Laplace equation in centeral difference form.
    Dirchlet boundary conditions are included
    '''
    A = np.zeros((N, N))
    b = np.zeros(N)
    A[0, 0] = 1.0
    for i in range(1, N-1):
        A[i, i-1:i+2] = [1.0, -2.0, 1.0]
    A[N-1, N-1] = 1.0
    return A, b
```

```python
def InitValue(N, k):
    '''
    Return initial guess of v => v0 = sin(i*k*pi/N), i=0~N
    '''
    return np.array([np.sin(float(i * k) * np.pi / float(N-1))
                    for i in range(N)])
```

We may also want to have a function that can return L2 norm of a vector.

```python
def L2norm(V):
    return np.sqrt(np.sum(V**2))
```

There are a minimum and a maximum value of wavenumber because the wave lengths are limited by spacing between grid points, $\Delta x$, and the length of computational domain, $L$: the minimum wave length is $\Delta x$, while the maximum is $2L$. If we define wavenumber $k$ as the numbers of half waves, then the maximum wavenumber is $N_s$, where $N_s = L/\Delta x$, and the minimum is $1$. So we can only pick wavenumbers between $1$ and $N_s$.


Let us discretize the space into $64$ segements (i.e. $N_s=64$), and the number of grid points, including end points, is $65$ ($N_p=65$).  We use three wavenumbers: $v_{initial}=sin(\pi x)$, $sin(3\pi x)$, and $sin(9\pi x)$, as shown below:

```python
# Initialize the number of grid points and coordinates
Np = 65
x = np.linspace(0., 1., Np)

# Initialize the wavenumbers
k = [1, 3, 9]

# Calculate initial guess
v0 = {ki: InitValue(Np, ki) for ki in k}

# Plot initial guess
ltype = ['k-', 'k--', 'k-.']  # line styles
plt.figure(figsize=(10, 5))
for i, ki in enumerate(k): plt.plot(x, v0[ki], ltype[i], label=r'$k='+str(ki)+'$');
plt.ylim((-1.1, 1.1));
plt.title(r'$v_{initial}=sin(k\pi x)$'); plt.xlabel(r'$x$'); plt.ylabel(r'$v_{initial}$')
plt.legend(bbox_to_anchor=(1.2, 0.25), fontsize='x-large', prop={'weight': 'bold'});
```

Now, we can start our computing. 

```python
iMethod = ['WJ', 'RB']  # WJ: Weighted Jacobi; RB: Red-Black Gauss-Seide
iFunc = {'WJ': WJacobi, 'RB': RBGaussSeidel}
iTitles = {'WJ': r'$Weighted\ Jacobi\ Method,\ \omega=2/3$',
           'RB': r'$Red-Black\ Gauss-Seidel\ Method$'}

# Initialize the coefficient matrix A, and constant vector b
A, b = Init1DLaplaceAb(Np)

# Set the number of iterations
Niter = 100

# Initialize the error vectors
err = {m: {ki: np.empty(Niter+1) for ki in k} for m in iMethod}

# Start to iterate 100 times for different wavenumber k
for m in iMethod:
    for ki in k:
        v = v0[ki].copy() # v: the approximated soln. vector
        err[m][ki][0] = L2norm(v)
        for i in range(1, Niter+1):
            v = iFunc[m](Np, A, v.copy(), b)
            err[m][ki][i] = L2norm(v)        
        err[m][ki] = err[m][ki] / np.amax(err[m][ki])
```

```python
# Plots
text_pos = int(Niter * 0.8)
plt.figure(figsize=(12, 6))
for mi, m in enumerate(iMethod):
    plt.subplot(1, 2, mi+1)
    plt.xlabel('$Iterations$')
    plt.ylabel('$Normalized\ L2\ Norm\ of\ Error$')
    plt.title(iTitles[m])
    plt.ylim((1e-9, 1e1))
    for i, ki in enumerate(k):
        plt.semilogy(range(Niter+1), err[m][ki], ltype[i], lw=2, label=r'$k='+str(ki)+'$')
        plt.legend(loc=0)    
plt.tight_layout()
```

We can see that the higher-wavenumber waves can be eliminated faster! 

Real errors are combinations of waves with defferent wavenumbers. This means that, because iterative methods can only eliminate higher-wavenumber waves efficiently but leave lower-wavenumber waves almost unchanged, the convergence of real errors will get slower or even get stuck after higher-wavenumber waves converged. Let's check this now using an initial guess which is a combination of three waves:

$$
v_{initial}=[sin(\pi x)+sin(6\pi x)+sin(32\pi x)]/3
$$

```python
# Calculate initial guess
v0 = (InitValue(Np, 1.0) + InitValue(Np, 6.0) + InitValue(Np, 32.0)) / 3.0

# Plot initial guess
plt.figure(figsize=(8, 4))
plt.plot(x, v0, 'k-')
plt.ylim((-1.1, 1.1))
plt.title(r'$v_{initial}=[sin(\pi x)+sin(6\pi x)+sin(32\pi x)]/3$')
plt.xlabel(r'$x$'); plt.ylabel(r'$v_{initial}$');
```

The contribution of each wave can be observed from the above figure.

```python
# Initialize the error vectors
err = {m: np.empty(Niter+1) for m in iMethod}

# for saving approximated soln at intermediate steps
transIter = [1, 5, 10, 20, 50, 100]
transV = {m: np.empty(6, dtype=np.ndarray) for m in iMethod}

for m in iMethod:
    c = 0
    v = v0.copy()
    err[m][0] = L2norm(v)
    # Start to iterate 100 times for different wavenumber k
    for i in range(1, Niter+1):
        v = iFunc[m](Np, A, v.copy(), b)
        err[m][i] = L2norm(v)
        if i == transIter[c]:
            transV[m][c] = v.copy()
            c += 1
    err[m] = err[m] / np.amax(err[m])
```

The approximated solutions at intermediate iterations are as showns below:

```python
# Plot intermediate solutions
ltype = {'WJ': 'k-', 'RB': 'b--'}
plt.figure(figsize=(12, 8))
for m in iMethod:
    for i in range(6):
        plt.subplot(3, 2, i+1)
        plt.plot(x, transV[m][i], ltype[m])
        plt.ylim((-1.0, 1.0))
        plt.xlabel(r'$x$'); plt.ylabel(r'$v$')
        plt.title(r'$Iteration\ \#'+str(transIter[i])+'$')
plt.legend(iTitles.values(), bbox_to_anchor=(0.85, 5), ncol=2, fontsize='xx-large')
plt.tight_layout()
```

The above set of figures shows that the highest-wavenumber wave is eliminated immediately within 5 iterations, and the second-high-wavenumber wave is eliminated in 20 iterations. The lowest-wavenumber wave, however, still remains and can be seen obviously after 100 iterations.

The rate of convergence is also examined:

```python
# Plot
plt.figure(figsize=(7,5))
plt.title(r'$Convergence\ of\ Errors$' + '\n' + 
          r'$v_{initial}=[sin(\pi x)+sin(6\pi x)+sin(32\pi x)]/3$')
for m in iMethod:
    plt.semilogy(range(Niter+1), err[m], ltype[m], label=iTitles[m])

plt.ylim((1e-1, 1e0))
plt.xlabel(r'$Iterations$')
plt.ylabel(r'$Normalized\ L_2\ Norm\ of\ Error$')
plt.legend(loc=0)
plt.tight_layout();
```

The above figure indeed shows a much slower convergence rate after a rapid convergence at the beginning. The rapid convergence represents the efficient elimination of the high-wavenumber wave; after the high-wavenumber wave is totally eliminated, the convergence rate slows down.

We can thus say that, if we want iterative methods to be more efficient, we should try to figure out a way to accelerate the eliminations of low-wavenumber waves.

Now let's see how a wave converges on grids with different resolutions. First, we examine how waves on a fine grid look like after we map it to a coarse grid where $N_{s,coarse}=N_{s,fine}/2$:

```python
GridName = ["Coarse", "Fine"]
Ns = {"Coarse": 16, "Fine": 32}
Np = {GN: Ns[GN]+1 for GN in GridName}

# Set up x coordinates
x = {GN: np.linspace(0., 1., Np[GN]) for GN in GridName}

# Set up the wavenumbers
k = [1, 8, 16, 21, 26, 32]

plt.figure(figsize=(18, 8), dpi=100)
ltype = {"Coarse": "k-", "Fine": "r--"}
for GN in GridName:
    for i, ki in enumerate(k):
        y = InitValue(Np[GN], ki)

        plt.subplot(2, 3, i+1)
        plt.ylim((-1.0, 1.0))
        plt.plot(x[GN], y, ltype[GN], label=r'$' + GN + ',\ N_s=' + str(Ns[GN]) + '$')
        plt.title(r'$k=' + str(ki) + '$')
plt.legend(ncol=2, bbox_to_anchor=(-0.75, 2.65), fontsize=28, frameon=True, fancybox=True)
plt.tight_layout()
```

From the above figures, we can see:

1. The waves with $k<N_{s,fine}/2$ (i.e. $k<N_{s,coarse}$) keep their wavenumbers on the coarse grid.
2. The waves with $k>N_{s,fine}/2$ (i.e. $k>N_{s,coarse}$) become waves with $k<N_{s,fine}/2$ (i.e. $k<N_{s,coarse}$) on the coarse grid. This phenomenon is called [***Aliasing***](http://nbviewer.ipython.org/github/ketch/teaching-numerics-with-notebooks/blob/master/Aliasing.ipynb).
3. The wave with $k=N_{s,fine}/2$ becomes zero on the coarse grid, because the grid points of the coarse grid locate on roots of the wave.
4. The wave with $k=N_{s,fine}$ becomes zero on both grids, because the grid points of both grids locate on roots of the wave.

Before we continue discussion, two terms are defined here: on a given grid, waves/errors with wavenumber greater than $N_s/2$ is called ***high-frequency*** (or ***oscillatory***) ***waves/errors***; waves/errors with wavenumber lower than $N_s/2$ is called ***low-frequency*** (or ***smooth***) ***waves/errors***.

Though the wavenumbers remain unchanged, the smooth waves of the fine grid become oscillatory waves of coarse grid, because the maximum wavenumber of the coarse grid is smaller than that of the fine grid. We also know from previous discussion that, on a given grid, more oscillatory waves converge faster. Then the question now is: will the smooth waves of the fine grid converge faster if we map the wave to and perform the calculation on the coarse grid?  The convergence rates of smooth waves of the fine grid on both fine and coarse grid are as shown below:  

```python
# Initialize coefficient matrix A and constant vector b
A = {}; b = {}
for GN in GridName:
    A[GN], b[GN] = Init1DLaplaceAb(Np[GN])

# Initial guess and wavenumber k
k = [5., 10., 15.]
v0 = {GN: {ki: InitValue(Np[GN], ki) for ki in k} for GN in GridName}

# Initialize the error vector
err = {GN: {ki: np.empty(Niter+1) for ki in k} for GN in GridName}

# Start to iterate 100 times for different grids
for GN in GridName:
    for ki in k:
        v = v0[GN][ki].copy()
        err[GN][ki][0] = L2norm(v)
        for i in range(1, Niter+1):
            v = WJacobi(Np[GN], A[GN], v.copy(), b[GN])
            err[GN][ki][i] = L2norm(v)

        err[GN][ki] = err[GN][ki] / np.amax(err[GN][ki])
```

```python
# Plot
plt.figure(figsize=(6*len(k), 6))
for i, ki in enumerate(k):
    plt.subplot(1, len(k), i+1)
    for GN in GridName:
        plt.semilogy(range(Niter+1), err[GN][ki], ltype[GN], 
                     label=r'$'+GN+'\ Grid,\ N_s=' + str(Ns[GN]) + '$')
    plt.title(r'$Convergence\ of\ Errors$' + '\n' +
              r'$of\ Different\ Grid\ Resolutions$' + '\n' +
              r'$k=' + str(ki) + '$')
    plt.ylim((1e-16, 1e0))
    plt.xlabel(r'$Iterations$')
    plt.ylabel(r'$Normalized\ L_2\ Norm\ of\ Error$')
    plt.legend(loc=0)
plt.tight_layout();
```

Yes! The smooth waves of the fine grid converge faster on the coarse grid!!

Here we only show the smooth waves of the fine grid, because this is what we care about in earlier discussion -- we want to figure out a way to efficiently eliminate the smooth errors (waves). So now the concept of multigrid methods finally emerges: we can eliminate oscillatory errors on a fine grid, map the remaining errors, which are smooth errors of the fine grid, to a coarse grid, and then eliminate these smooth errors on the coarse grid! In fact, the multigrid methods can use more than 2 grids. After the convergence gets stuck on the coarse grid, we can again map the remaining errors to a much coarser grid. This mapping procedure can be performed again and again until it reaches some specific condiction, e.g. the grid resolution reaches the coarsest state.
