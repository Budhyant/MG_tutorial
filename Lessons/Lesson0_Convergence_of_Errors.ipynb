# Lesson 0: Convergence of Errors
-----------------------  
Existing numerical methods for solving a linear system of equations, $Au=f$, fall into two large categories: *direct methods* and *iterative (or relaxation) methods*. Iterative methods show an interesting property for the convergence of errors: the convergence rates depend on the frequencies of the errors. This interesting property is the basis of multigrid methods.  We will start this multigrid tutorial with an introduction of iterative methods and a discussion of error.

We first review classical iterative methods. In the second section, we define the example problem we will use to examine the convergence of errors. Finally, we discuss the convergence of errors in the third section.

## 0-1 Basic Iterative Methods
Iterative methods begin with an initial guess at a solution and try to improve the approximated solution through iterations. Ideally, the approximated solutions generated from each iteration should converge to the exact solution. 

We review two methods here -- the weighted Jacobi method and the Red-Black Gauss-Seidel method.

### 0-1-1 Weighted Jacobi Method
If $\mathbf{v}^{(m)}$ represents the approximated solutions at $m_{th}$ iteration, $\mathbf{v}^{(m+1)}$ in Jacobi Method is:

$$
v{_i^{(m+1)}} = [b_i - (A_{i,0}v{_0^{(m)}}+\cdots+A_{i,i-1}v{_{i-1}^{(m)}}+A_{i,i+1}v{_{i+1}^{(m)}}+\cdots+A_{i,n}v{_n^{(m)}})] / A_{i,i}
$$

Or we can express it in matrix form:

$$
\mathbf{v}^{(m+1)}=D^{-1}(\mathbf{b} - (L+U) \mathbf{v}^{(m)})
$$

where $D$, $L$, and $U$ represent the diagonal, lower triangular, and upper triangular parts of $A$ respectively, i.e. $A = D + L + U$. 

The above matrix equation can also be written as:

$$
\mathbf{v}^{(m+1)}=\mathbf{v}^{(m)} + D^{-1}(\mathbf{b} - A\mathbf{v}^{(m)})
$$

From this matrix equation, we can see that the new approximated solution $\mathbf{v}^{(m+1)}$ is actually the old approximated solution $\mathbf{v}^{(m)}$ plus a correction term ($D^{-1}(\mathbf{b} - A\mathbf{v}^{(m)})$). The weighted Jacobi method tries to adjust the amount of correction by defining a parameter $\omega$ such that:

$$
\mathbf{v}^{(m+1)}=\mathbf{v}^{(m)} + \omega D^{-1}(\mathbf{b} - A\mathbf{v}^{(m)})
$$  
If $\omega=1$, the weighted Jacobi method is equivalent to the traditional Jacobi method. 

The Python code for weighted Jacobi method is:

```python
def WJacobi(N, A, v0, b, omg, Niter):
    '''
    Inputs:
        N: the dimension of A, v0, and b (integer)
        A: coefficient matrix (N by N array)
        v0: initial guess (N by 1 array)
        b: constant matrix (N by 1 array)
        omg: relaxation parameter (float)
        Niter: the number of iterations to perform
    Return:
        The approximated solution v after Niter iterations
        (N by 1 array)
    '''
    v = np.empty(N)
    DA = np.diag(A)
    for ITER in range(Niter):
        v = v0 + omg * (b - np.dot(A, v0)) / DA
        v0 = v.copy()
    return v
```

### 0-1-2 Red-Black Gauss-Seidel Method
In the weighted Jacobi method, each element in the new approximated solution $\mathbf{v}^{(m+1)}$ is constructed from elements in previous approximated solution $\mathbf{v}^{(m)}$ only. What if, instead, we use elements from the new approximated solution to calculate other elements that we are updating? That is

$$
v{_i^{(m+1)}} = [b_i - (A_{i,0}v{_0^{(m+1)}}+\cdots+A_{i,i-1}v{_{i-1}^{(m+1)}}+A_{i,i+1}v{_{i+1}^{(m)}}+\cdots+A_{i,n}v{_n^{(m)}})] / A_{i,i}
$$

This is called the Gauss-Seidel method. It can converge faster than the weighted Jacobi method, but the computation time required may still exceed that of the weighted Jacobi method as the Gauss-Seidel method is hard to parallelize.

In fact, though the equation shown above updates elements consecutively from the first to last, the order can be changed. One of the modified Gauss-Seidel methods is called the red-black Gauss-Seidel method: it updates the even elements first and then updates the odd elements, as shown below.

$$
\text{step }1:~~ v{_{2i}^{(m+1)}} = [b_{2i} - (\cdots+A_{2i,2i-2}v{_{2i-2}^{(m)}}+A_{2i,2i-1}v{_{2i-1}^{(m)}}+A_{2i,2i+1}v{_{2i+1}^{(m)}}+A_{2i,2i+2}v{_{2i+2}^{(m)}}+\cdots)] / A_{2i,2i}
$$

$$
\text{step }2:~~ v{_{2i+1}^{(m+1)}} = [b_{2i+1} - (\cdots+A_{2i,2i-2}v{_{2i-1}^{(m)}}+A_{2i,2i}v{_{2i}^{(m+1)}}+A_{2i,2i+2}v{_{2i+2}^{(m+1)}}+A_{2i,2i+3}v{_{2i+3}^{(m)}}+\cdots)] / A_{2i+1,2i+1}
$$

It only uses information from the previous approximation when updating the even elements in the first step and then uses updated information from the first step when updating the odd elements. Both step $1$ and step $2$ can be parallelized individually.

The code for the red-black Gauss-Seidel method is:

```python
def RBGaussSeidel(N, A, v0, b, Niter):
    '''
    Inputs:
        N: the dimension of A, v0, and b (integer)
        A: coefficient matrix (N by N array)
        v0: initial guess (N by 1 array)
        b: constant matrix (N by 1 array)
        Niter: the number of iterations to perform
    Return:
        The approximated solution v after Niter iterations
        (N by 1 array)
    '''
    DA = np.diag(A)
    for ITER in range(Niter):
        v0[0::2] = v0[0::2] + \
            (b[0::2] - np.dot(A[0::2, :], v0)) / DA[0::2]
        v0[1::2] = v0[1::2] + \
            (b[1::2] - np.dot(A[1::2, :], v0)) / DA[1::2]
    return v0
```

## 0-2 Example Problems
Before we start to discuss the convergence of errors of iterative methods, we introduce the example problem which will be used later. Consider the following ODE:

$$-u''(x)+\sigma u(x)=f(x),~~0<x<1,~~\sigma\ge0$$  
$$u(0)=u(1)=0$$

We discretize the domain into $N_s$ segments, so the number of grid points, $N_p$, is $N_s+1$, including end points. Applying central difference on the above ODE, we can obtain a linear system (including the boundary conditions): $A\mathbf{u}=\mathbf{b}$, where the vector $\mathbf{u}=(u_0,...,u_n)$ represents the exact solution on each point, and 

$$
A=
\begin{bmatrix}
1 & 0 & 0 & \cdots & \cdots & 0 \\
-1 & 2+\sigma \Delta x^2 & -1 & \ddots & &\vdots \\
\vdots & \ddots & \ddots & \ddots & \ddots & \vdots \\
\vdots & \ddots & \ddots & \ddots & \ddots & \vdots\\
\vdots & & \ddots & -1 & 2+\sigma \Delta x^2 & -1\\
0 & \cdots & \cdots & 0 & 0 & 1
\end{bmatrix},~~
\mathbf{b}=
\Delta x^2
\begin{bmatrix}
0 \\ f_1 \\ \vdots \\ \vdots \\ f_{n-1} \\ 0
\end{bmatrix}
$$

For convenience, we will assume $\sigma=0$ and $f(x)=0$ (i.e. 1D Laplace equation) later in this notebook. The exact solution of this Laplace equation is $u(x)=0$ (i.e. $\mathbf{u}=(0,...,0)$). The vector $\mathbf{v}^{(m)}$ still represents the approximated solution at $m_{th}$ iteration in iterative methods.

## 0-3 The Convergence of Errors of Iterative Methods
For a given approximated solution $\mathbf{v}^{(m)}$, the error vector $\mathbf{e}^{(m)}$ is simply $\mathbf{e}^{(m)}=\mathbf{v}^{(m)}$ in the 1D Laplace equation. The error can be presented by Fourier series, which decomposes the error into a combination of waves with different wavenumbers or frequencies. When error converges to zero, it means all the waves also converge to zero. If the convergence rates of the waves are not a constant and depends on the wavenumbers, then the convergence rate of the error depends on the waves that converge most slowly. 

In order to find out the relationships between the convergence rates and the wavenumbers, we make the initial guess, $v_{initial}$, to be a wave with a single wavenumber. As mentioned in the last paragraph, the approximated solution is right the error in 1D Laplace equation. This means now our initial error, $e_{initial}$, only contains a single wavenumber. Let's see how fast the iterative methods eliminate errors with different wavenumbers. 

First, for convenience, we write the initializations into functions.

```python
def Init1DLaplaceAb(N):
    '''
    Initialize matrix A and vector b
    for the 1D Laplace equation in centeral difference form.
    Dirchlet boundary conditions are included
    '''
    A = np.zeros((N, N))
    b = np.zeros(N)
    A[0, 0] = 1.0
    for i in range(1, N-1):
        A[i, i-1:i+2] = [1.0, -2.0, 1.0]
    A[N-1, N-1] = 1.0
    return A, b
```

```python
def InitValue(N, k):
    '''
    Return initial guess of v => v0 = sin(i*k*pi/N), i=0~N
    '''
    return np.array([np.sin(float(i * k) * np.pi / float(N-1))
                    for i in range(N)])
```

We may also want to have a function that can return L2 norm of a vector.

```python
def L2norm(V):
    return np.sqrt(np.sum(V**2))
```

Now, we can start our computing. Let us discretize the space into $64$ segements (i.e. $N_s=64$), and the number of grid points, including end points, is $65$ ($N_p=65$).

```python
# Initialize the grid points, coefficient matrix, and constant matrix
Np = 65
A, b = Init1DLaplaceAb(Np)

# Set the number of iterations
Niter = 100
ITERS = np.array(range(Niter+1))

# Initialize the approximated soln. vectors and error vectors
v = np.empty((2, Np))
err = np.empty((3, 2, Niter+1))

# Initialize the wavenumbers
k = [1, 3, 9]
# Start to iterate 100 times for different wavenumber k
for m in [0, 1, 2]:
    
    for i in [0, 1]:
        v[i] = InitValue(Np, k[m])
        err[m, i, 0] = L2norm(v[i])

    for ITER in ITERS[1:]:
        v[0] = WJacobi(Np, A, v[0].copy(), b, 2./3., 1)
        v[1] = RBGaussSeidel(Np, A, v[1].copy(), b, 1)
        for i in [0, 1]:
            err[m, i, ITER] = L2norm(v[i])

    for i in [0, 1]:
        err[m, i, :] = err[m, i, :] / np.amax(err[m, i, :])
```

```python
# figure
fig1, axs = plt.subplots(1, 2, figsize=(10, 5), dpi=100)
STR_POS = int(Niter * 0.8)
titles = [r'Weighted Jacobi Method with $\omega=\frac{2}{3}$',
          'Red-Black Gauss-Seidel Method']
for i in [0, 1]:
    axs[i].grid()
    axs[i].set_xlabel('Iterations')
    axs[i].set_ylabel('Normalized L2 Norm of Error')
    axs[i].set_title(titles[i])
    for m in [0, 1, 2]:
        axs[i].plot(ITERS, err[m, i, :], 'k-', lw=2)
        axs[i].text(STR_POS, err[m, i, STR_POS]+0.01, 'k='+str(k[m]))
        axs[i].legend([r'$v_{initial}=sin(k\pi x)$'], loc=0)

fig1.tight_layout(h_pad=3)    
```

We can see that the higher-wavenumber waves can be eliminated faster. Real errors are combinations of waves with defferent wavenumbers. This means iterative methods can only eliminate higher-wavenumber waves efficiently but leave lower-wavenumber waves almost unchanged. So the convergence of real errors will get slower or even get stuck after higher-wavenumber waves converged. Let's check this now using an initial guess with mixture waves.

```python
# Initialize the approximated soln. vectors and error vectors
v = np.empty((2, Np))
err = np.empty((2, Niter+1))
for i in [0, 1]:
    v[i] = (InitValue(Np, 1.0) + InitValue(Np, 6.0) + InitValue(Np, 32.0)) / 3.0
    err[i, 0] = L2norm(v[i])
    
# Start to iterate 100 times for different wavenumber k
for ITER in ITERS[1:]:
    v[0] = WJacobi(Np, A, v[0].copy(), b, 2./3., 1)
    v[1] = RBGaussSeidel(Np, A, v[1].copy(), b, 1)
    for i in [0, 1]:
        err[i, ITER] = L2norm(v[i])

for i in [0, 1]:
    err[i, :] = err[i, :] / np.amax(err[i, :])
```

```python
# Plot
fig2 = plt.figure(figsize=(5,5), dpi=100)
plt.title('Convergence of Errors with Mixture Wavenumbers')
plt.plot(ITERS, err[0, :], 'k-', lw=2, label=titles[0])
plt.plot(ITERS, err[1, :], 'k--', lw=2, label=titles[1])
plt.ylim((0, 1.0))
plt.xlabel('Iterations')
plt.ylabel('Normalized L2 Norm of Error')
plt.grid()
plt.legend(loc=0)
plt.text(5, 0.3, 
         r'$v_{initial}=[sin(\pi x)+sin(6\pi x)+sin(32\pi x)]/3$',
         fontsize=14)
```

The above figure indeed shows a much slower convergence rate after a rapid convergence at the beginning. The rapid convergence represents the efficient elimination of the high-wavenumber wave; after the high-wavenumber wave is totally eliminated, the convergence rate slows down.

We can thus say that, if we want iterative methods to be more efficient, we should try to figure out a way to accelerate the eliminations of low-wavenumber waves.

Now let's see how a wave converges on grids with different numbers of gird points.

```python
# Three different resolutions are used
Ns = np.array([16, 32, 64])
Np = Ns + 1

A = []
b = []
v = [[], []]
k = [1., 63.]
err = np.empty((2, 3, Niter+1))

for i in range(3):
    A.append(np.empty((Np[i], Np[i])))
    b.append(np.empty(Np[i]))
    A[i], b[i] = Init1DLaplaceAb(Np[i])
    v[0].append(InitValue(Np[i], k[0]))
    v[1].append(InitValue(Np[i], k[1]))
    err[0, i, 0] = L2norm(v[0][i])
    err[1, i, 0] = L2norm(v[1][i])
    
# Start to iterate 100 times for different wavenumber k
for ITER in ITERS[1:]:
    for i in range(3):
        v[0][i] = WJacobi(Np[i], A[i], v[0][i].copy(), b[i], 2./3., 1)
        v[1][i] = WJacobi(Np[i], A[i], v[1][i].copy(), b[i], 2./3., 1)
        err[0, i, ITER] = L2norm(v[0][i])
        err[1, i, ITER] = L2norm(v[1][i])

for i in range(3):
    err[0, i, :] = err[0, i, :] / np.amax(err[0, i, :])
    err[1, i, :] = err[1, i, :] / np.amax(err[1, i, :])
```

```python
# Plot
fig3 = plt.figure(figsize=(10, 5), dpi=100)
for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.title('Convergence of Errors \n' + 
              'Using Different Number of Grid Points \n' +
              '(' + titles[0] + ')')
    plt.plot(ITERS, err[i, 0, :], 'k-', lw=2, label=r'$N_p=$'+str(Np[0]))
    plt.plot(ITERS, err[i, 1, :], 'k--', lw=2, label=r'$N_p=$'+str(Np[1]))
    plt.plot(ITERS, err[i, 2, :], 'k-.', lw=2, label=r'$N_p=$'+str(Np[2]))
    plt.ylim((0, 1.0))
    plt.xlabel('Iterations')
    plt.ylabel('Normalized L2 Norm of Error')
    plt.grid()
    plt.legend(loc=0)
    plt.text(10, 0.4, 
             r'$v_{initial}=sin($' + str(k[i]) + r'$\pi x)$',
             fontsize=14)
```

This is really interesting! The convergence rate of lowest-wavenumber wave gets faster on coarser grids, while it's not exactly true for the second-highest-wavenumber wave!

When we decompose the errors into waves on a discretized space, the wavenumer $k$ is descritized and has minimum and maximum values, because the wave lengths are limited by spacing between grid points, $\Delta x$, and the length of computational domain, $L$. The minimum wave length is $\Delta x$, while the maximum is $2L$. Thus the maximum wavenumber is $N_s$, and the minimum is $1$. Therefore, in the above example, the maximum wavenumber in the three cases are $k_{max}=16,~32,~64$ respectively. However, using the wave with highest wavenumber here is meaningless, since all points locate on interger multiples of $\pi$ in this case and the initial guess is a zero vector, which is the exact solution. So we use second-highest wavenumber on the $N_p=65$ grid in the above figure.

For a wave with a low wavenumber, $k=1$ for example, this wave in $N_p=17$ case is a relatively higher wavenumber than it is in $N_p=33$ case, as the maximum wavenumber in previous case is only $17$, while that in later case is $33$. And the same concept can be applied on $N_p=33$ and $N_p=65$ cases. So it looks reasonable that the lower-wavenumber waves can be eliminated more efficiently on coarse grids.

But how about $k=63$? We can check what a wave with $k=63$ looks like in the following figure.

```python
x1 = np.linspace(0., 1., Np[2])
x2 = np.linspace(0., 1., Np[1])
x3 = np.linspace(0., 1., Np[0])

plt.figure(figsize=(10, 5), dpi=100)
plt.plot(x1, np.sin(63. * np.pi * x1), 'k-', lw=2, label=r'$N_p=$'+str(Np[2]))
plt.plot(x2, np.sin(63. * np.pi * x2), 'r--', lw=3, label=r'$N_p=$'+str(Np[1]))
plt.plot(x3, np.sin(63. * np.pi * x3), 'b-.', lw=3, label=r'$N_p=$'+str(Np[0]))
plt.ylim((-1., 1.0))
plt.grid()
plt.legend(loc=0)
```

The $k=63$ wave on $N_p=65$ grid becomes $k=1$ waves on both $N_p=33$ and $N_p=17$ grids! This illustrates the convergence rates of $k=63$ wave on different grids on the previous figure.

We can check deeper by more figures:

```python
Np1 = 16 + 1
Np2 = 32 + 1
x1 = np.linspace(0., 1., Np1)
x2 = np.linspace(0., 1., Np2)

k = [2, 9, 16, 25, 31]

plt.figure(figsize=(18, 8), dpi=100)

for i in range(5):
    y1 = InitValue(Np1, k[i])
    y2 = InitValue(Np2, k[i])

    plt.subplot(2, 3, i+1)
    plt.ylim((-1.0, 1.0))
    plt.plot(x1, y1, 'k-', lw=2, label=r'$N_p=17$')
    plt.plot(x2, y2, 'r--', lw=2, label=r'$N_p=33$')
    plt.title('k=' + str(k[i]))
    plt.legend(loc=0)
```

The above figures show:

1. Waves with wavenumbers lower than $N_{s,fine}/2$ on the fine grid become relatively high-wavenumber waves on the coarse grid.
2. When the wave on the fine grid has wavenumber equal to $N_{s, fine}/2$, it will return a $0$ vector on the coarse grid.
3. Waves with wavenumbers higher than $N_{s, fine}/2$ on the fine grids become low-wavenumber waves on the coarse grid. (This is because in the coarse gird, the wavenumbers can not be greater than $N_{s, coarse}$, and the wave with wavenumber higher than $N_{s, coarse}$ will be represented by lower-wavenumber waves. This phenomenon is called *aliasing*.)

We define two terms here for waves with wavenumbers lower than $N_s /2$ and that higher than $N_s /2$: low-frequency (or smooth) modes and high-frequency (or oscillatory) modes. These terms come from the fact that these waves are the Fourier modes of errors: we can decompose the errors into different waves by Fourier series expansion.

Combining the above observations and the fact that high-frequency modes converge faster, the basic concept of multigrid methods emerges: we can eliminate high-frequency modes on a fine grid ,then map low-frequency modes to a coarse grid, and eliminate these low-frequency modes on the coarse grid.
