```python
def C2F1D(NC, vC):
    '''
    Map a vector on a coarse grids to a fine grid by linear interpolation.
    
    Input:
        NC: the number of elements of the input vector (integer)
        vC: the vector on the coarse grid (1D array)
    Output:
        NF: the number of elements of the output vector (integer)
        vF: the vector on the fine grid (1D array)
    '''
    NF = 2 * NC - 1
    vF = np.empty(NF)
    for i in range(NC-1):
        vF[2*i] = vC[i]
        vF[2*i+1] = (vC[i] + vC[i+1]) * 0.5
    vF[-1] = vC[-1]
    return NF, vF
```

```python
GridName = ["Fine_Exact", "Fine_Mapped", "Coarse",]
Np = {"Coarse": 17, "Fine_Mapped": 33, "Fine_Exact": 33}

x = {GN: np.linspace(0., 1., Np[GN]) for GN in GridName} 

k = [5, 15]

ltype = {"Coarse": "r^--", "Fine_Mapped": "bo-.", "Fine_Exact": "k-"}
titles = {"Coarse": "Exact Solution on Coarse Grid", 
          "Fine_Mapped": "Solution on Fine Grid Mapped from Coarse Grid", 
          "Fine_Exact": "Exact Solution on Fine Grid"}

plt.figure(figsize=(18, 8))
for i, ki in enumerate(k):
    
    v = {"Coarse": InitValue(Np["Coarse"], ki), 
         "Fine_Exact": InitValue(Np["Fine_Exact"], ki)} 
    v["Fine_Mapped"] = C2F1D(Np["Coarse"], v["Coarse"])[1]
    
    plt.subplot(1, 2, i+1)
    for GN in GridName:
        plt.plot(x[GN], v[GN], ltype[GN], markersize=10, label=titles[GN])

    plt.title(r"Wave of $k=" + str(ki) + "$ on Coarse and Fine Grid")
    plt.ylim((-1.1, 1.5)); plt.xlabel(r"$x$"); plt.ylabel(r"$v$")
    plt.legend(loc=0, numpoints=1);
```

```python
def F2C1D(NF, vF):
    '''
    Map a vector on a fine grids to a coarse grid by linear interpolation.
    
    Input:
        NF: the number of elements of the input vector (integer)
        vF: the vector on the fine grid (1D array)
    Output:
        NC: the number of elements of the output vector (integer)
        vC: the vector on the fine grid (1D array)
    '''
    NC = (NF + 1) / 2
    vC = np.empty(NC)
    vC[0] = vF[0]
    for i in range(1, NC-1):
        vC[i] = 0.25 * (vF[2*i-1] + 2. * vF[2*i] + vF[2*i+1])
    vC[-1] = vF[-1]
    return NC, vC
```

```python
GridName = ["Coarse_Exact", "Coarse_Mapped", "Fine",]
Np = {"Coarse_Exact": 17, "Coarse_Mapped": 17, "Fine": 33}

x = {GN: np.linspace(0., 1., Np[GN]) for GN in GridName} 

k = [5, 15]

ltype = {"Fine": "r^--", "Coarse_Mapped": "bo-.", "Coarse_Exact": "k-"}
titles = {"Coarse_Exact": "Exact Solution on Coarse Grid", 
          "Coarse_Mapped": "Solution on Coarse Grid Mapped from Fine Grid", 
          "Fine": "Exact Solution on Fine Grid"}

plt.figure(figsize=(18, 8))
for i, ki in enumerate(k):
    
    v = {"Coarse_Exact": InitValue(Np["Coarse_Exact"], ki), 
         "Fine": InitValue(Np["Fine"], ki)} 
    v["Coarse_Mapped"] = F2C1D(Np["Fine"], v["Fine"])[1]
    
    plt.subplot(1, 2, i+1)
    for GN in GridName:
        plt.plot(x[GN], v[GN], ltype[GN], markersize=10, label=titles[GN])

    plt.title(r"Wave of $k=" + str(ki) + "$ on Coarse and Fine Grid")
    plt.ylim((-1.1, 1.5)); plt.xlabel(r"$x$"); plt.ylabel(r"$v$")
    plt.legend(loc=0, numpoints=1);
```
