```python
def C2F1D(NC, vC):
    '''
    Map a vector on a coarse grids to a fine grid by linear interpolation.
    
    Input:
        NC: the number of elements of the input vector (integer)
        vC: the vector on the coarse grid (1D array)
    Output:
        NF: the number of elements of the output vector (integer)
        vF: the vector on the fine grid (1D array)
    '''
    NF = 2 * NC - 1
    vF = np.empty(NF)
    for i in range(NC-1):
        vF[2*i] = vC[i]
        vF[2*i+1] = (vC[i] + vC[i+1]) * 0.5
    vF[-1] = vC[-1]
    return NF, vF
```

```python
N1 = 5; N2 = 9

x1 = np.linspace(0., 1., N1); x2 = np.linspace(0., 1., N2)
x_exact = np.linspace(0., 1., 101)

v1 = InitValue(N1, 3); v2 = C2F1D(N1, v1)[1]
v_exact = InitValue(101, 3)

plt.plot(x1, v1, 'r^--', lw=2, markersize=14, label='Values on Coarse Grid')
plt.plot(x2, v2, 'bo-.', lw=2, markersize=8, label='Values on Fine Grid')
plt.plot(x_exact, v_exact, 'k-', lw=2, label='Exact Solution')
plt.grid(); plt.legend(loc=0, fontsize="small", numpoints=1, markerscale=0.5)
```

```python
def F2C1D(NF, vF):
    '''
    Map a vector on a fine grids to a coarse grid by linear interpolation.
    
    Input:
        NF: the number of elements of the input vector (integer)
        vF: the vector on the fine grid (1D array)
    Output:
        NC: the number of elements of the output vector (integer)
        vC: the vector on the fine grid (1D array)
    '''
    NC = (NF + 1) / 2
    vC = np.empty(NC)
    vC[0] = vF[0]
    for i in range(1, NC-1):
        vC[i] = 0.25 * (vF[2*i-1] + 2. * vF[2*i] + vF[2*i+1])
    vC[-1] = vF[-1]
    return NC, vC
```
