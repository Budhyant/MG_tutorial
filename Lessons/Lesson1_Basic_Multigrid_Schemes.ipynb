# Lesson 1: Basic Multigrid Schemes
-----------------------  
As the conclusion at the end of Lesson 0, the aim of utilization of multigrids is to eliminate oscillatory errors in fine grids and eliminate smooth errors in coarse grids. The strategies that can achieve this typically fall into to two categories: 1) ***nested iteration*** and 2) ***correction scheme***. We will demonstrate these two scheme in this notebook.

## 1-1 Nested Iteration
The basic idea of nested iteration is to obtain better initial guesses by using coarse grids: make the initial guess on fine grids only contain oscillatory errors by eliminating the smooth errors in initial guess on coarse grids prior to that.

Before we continue on this stragety, first we have to define some notations. $\Omega^h$ is the original grid, which has the spacing between two grid points being $\Delta x=h$, and $\Omega^{mh}$ represents the grid of which the spacing is $m\times h$. In other words, if there are $N_p^h$ points (including end points) on $\Omega^h$, then there will be $(N_p^h-1)/m+1$ points on $\Omega^{mh}$, which can be denoted as $N_p^{2h}$.

In this discussion, we only consider the case in which the coarse gird has twice the grid spacing of the next fine grid. Then the basic idea of nested iteration can be represented as:

* Rleax on $Av=f$ on a very coarse grid to obtain an initial guess for the next grid.

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\vdots$

* Relax on $Av=f$ on $\Omega^{4h}$ to obtain an initial guess for $\Omega^{4h}$, and pass it to $\Omega^{2h}$.
* Relax on $Av=f$ on $\Omega^{2h}$ to obtain an initial guess for $\Omega^{2h}$, and pass it to $\Omega^{h}$.
* Relax on $Av=f$ on $\Omega^{h}$ to obtain a final approximation to the solution

This approach looks attractive. But the first question is: how to pass the initial guess from a coarser grid to a fine grid?

### 1-1-1 Mapping the Initial Guesses From Coarse Grids to Fine Grids
We can map the initial guess from a coarse grid to a fine grid through interpolation. Many interpolation methods can be used. For most multigrid purposes, the simplest way is linear interpolation. If $v_j^{2h}$ and $v_j^h$ represent the approximated solutions on $\Omega^{2h}$ and $\Omega^{h}$, then the linear interpolation is as shown below:
$$
v_{2j}^h=v_j^{2h}\text{,and}\\
v_{2j+1}^h=\frac{1}{2}\left(v_j^{2h}+v_{j+1}^{2h}\right)\text{, }0\le j\le \frac{N_p^h-1}{2}-1
$$
where $N_p^h$ is the number of points on $\Omega^h$. The index of points starts from $0$ in order to follow the convention of the numpy array in Python.

We write a Python function for this linear interpolation for convenience.

```python
# The following codes are written for Python 2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('../FigStyle/plot.mplstyle')
execfile("../Codes/IterativeMethods.py")
execfile("../Codes/Miscellaneous.py")
%matplotlib inline
```

```python
def C2F1D(NC, vC):
    '''
        Map a vector on a coarse grids to a fine grid by linear interpolation.
    
        Input:
            NC: the number of elements of the input vector (integer)
            vC: the vector on the coarse grid (1D array)
        
        Output:
            NF: the number of elements of the output vector (integer)
            vF: the vector on the fine grid (1D array)
    '''
    NF = 2 * NC - 1
    vF = np.empty(NF)
    for i in range(NC-1):
        vF[2*i] = vC[i]
        vF[2*i+1] = (vC[i] + vC[i+1]) * 0.5
    vF[-1] = vC[-1]
    return vF.copy()
```

Let's now see an example:

```python
Np = np.array([33, 33, 17])
x = np.array([np.linspace(0., 1., Np[i]) for i in range(3)],
             dtype=np.ndarray)
v = np.empty(3, dtype=np.ndarray)
k = np.array([5, 15])

ltype = np.array(["k-", "r^--", "bo-."])
labels = np.array(["Exact Solution on Fine Grid",
                   "Mapped Solution on Fine Grid",
                   "Approximated Solution on Coarse Grid"])

plt.figure(figsize=(14, 6))
for i, ki in enumerate(k): 
    v[[0, 2]] = [InitValue(Np[j], ki) for j in [0, 2]]
    v[1] = C2F1D(Np[2], v[2])
    
    plt.subplot(1, 2, i+1)
    for j in range(3):
        plt.plot(x[j], v[j], ltype[j], markersize=10, label=labels[j])
    
    plotAssist(title=r"Wave of $k=" + str(ki) + "$ on Coarse and Fine Grid", 
               ylmt=(-1.1, 1.5), legArg={"fontsize":'medium'})
```

The above two figures show the linear interpolations for a relatively smooth wave ($k=5$) and a relatively oscillatory wave ($k=15$). We can see that every thing looks fine for $k=5$. However, for $k=15$, when we map the approximated solution from the coarse grid to the fine grid, the mapped solution shows a deviation from the exact solution of the fine grid. This tells us that if the solution (or, generally speaking, a *function*) is oscillatory, then the linear interpolation will cause distortion on the wave. Therefore, in our multigrid procedure, this linear interpolation is most effective when the initial guess  (or the error) is smooth.

### 1-1-2 Mapping the Initial Guesses From Fine Grids to Coarse Grids
Now we have a new question. From the brief description of this nested iteration, it implies that we have to determine the initial guesses for those coarse girds. Mapping the initial guess on $\Omega^h$, which is the original grid we have for the problem, seems to be a good idea. We thus need methods to deal with this kind of operations. They are generally called *restriction*. In fact, this operation also gives us the idea of a V-cycle scheme for nested iteration, which we will talk about in the next section.

The most obvious method is *injection*. It is
$$
v_j^{2h}=v_{2j}^h
$$

In this notebook, we use another method called *full weighting*, which is defined as
$$
v_j^{2h}=\frac{1}{4}\left(v_{2j-1}^h+2v_{2j}^h+v_{2j+1}^h\right)\text{, } 1 \le j \le \frac{N_p-1}{2}-1
$$

The Python function for this restriction operation is:

```python
def F2C1D(NF, vF):
    '''
        Map a vector on a fine grids to a coarse grid by linear interpolation.
    
        Input:
            NF: the number of elements of the input vector (integer)
            vF: the vector on the fine grid (1D array)
        
        Output:
            NC: the number of elements of the output vector (integer)
            vC: the vector on the fine grid (1D array)
    '''
    NC = (NF + 1) / 2
    vC = np.empty(NC)
    vC[0] = vF[0]
    for i in range(1, NC-1):
        vC[i] = 0.25 * (vF[2*i-1] + 2. * vF[2*i] + vF[2*i+1])
    vC[-1] = vF[-1]
    return vC.copy()
```

An example of restriction is also provided here.

```python
Np = np.array([17, 33, 17])
x = np.array([np.linspace(0., 1., Np[i]) for i in range(3)], 
             dtype=np.ndarray)
v = np.empty(3, dtype=np.ndarray)
k = np.array([5, 15])

ltype = np.array(["k-", "r^--", "bo-."])
labels = np.array(["Exact Solution on Coarse Grid", 
                   "Approximated Solution on Fine Grid",
                   "Solution on Coarse Grid Mapped from Fine Grid"])

plt.figure(figsize=(14, 6))
for i, ki in enumerate(k):
    
    v[0:2] = [InitValue(Np[j], ki) for j in range(2)]
    v[2] = F2C1D(Np[1], v[1])
    
    plt.subplot(1, 2, i+1)
    for j in range(3):
        plt.plot(x[j], v[j], ltype[j], markersize=10, label=labels[j])

    plotAssist(title=r"Wave of $k=" + str(ki) + "$ on Coarse and Fine Grid", 
               ylmt=(-1.1, 1.5), legArg={"fontsize":'medium'})
```

The figure above show that this full weighting restriction lokks good on both waves!

### 1-1-3 V-Cycle Scheme
In the last section, we said we have to map the initial guess from finest grid to the coarsest grid before we actually relax $Av=f$ on the later one. Only mapping without solving other things seems to be wasting time. So the idea now is that why not solve $Av=f$ at the same step when we are mapping the initial guess from fine grids to coarse ones? Also, we can relax more than one time on each grids. The general description of this V-Cycle Scheme is described as below.


* Relax on $Av=f$ $n_1$ times on $\Omega^{h}$ to obtain an initial guess $\Omega^{h}$, and pass it to $\Omega^{2h}$.
* Relax on $Av=f$ $n_1$ times on $\Omega^{2h}$ to obtain an initial guess $\Omega^{2h}$, and pass it to $\Omega^{4h}$.

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\vdots$

* Rleax on $Av=f$ $n_2$ times on a very coarse grid to obtain an initial guess for the next grid.

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\vdots$

* Relax on $Av=f$ $n_2$ times on $\Omega^{2h}$ to obtain an initial guess for $\Omega^{2h}$, and pass it to $\Omega^{h}$.
* Relax on $Av=f$ $n_2$ times on $\Omega^{h}$ to obtain a final approximation to the solution

The relaxing on fine grid before mapping the initial guess to coarse grid can eliminate oscillatory errors first. In Lesson 0, we know *aliasing* will make oscillatory errors on fine grid to be more smooth on the coarse grid after restriction. This aliasing may increase the "loading" of the relaxing on the coarse grid. Therefore, V-Cycle scheme may be a good idea.

V-cycle scheme is not the only one way to realize the nested iteration. However, it is the basic one.

### 1-1-4 Example
Let's see how the abovementioned V-cycle nested iteration scheme performs and compare it to the original weighted Jacobi method. In this example, the initial guess on the finest grid is:

$$
v_0^h=\left[\sin(\pi x) + \sin(15\pi x) + \sin(31\pi x) +  \sin(63\pi x)\right]/4
$$

We use 3 grids to demonstrate this.

```python
# set up the number of grid points
Nph = 65; 

# set the number of grids
Ng = 3

# set up the number of grid points for all grids
Np = np.array([(Nph-1)/(2**i)+1 for i in range(Ng)])

# set up coordinates
x = np.array([np.linspace(0., 1., Np[i]) for i in range(Ng)], 
             dtype=np.ndarray)

# set up initial guess for the finest grid
# v0 = 2 * np.random.random(Nph) - 1
v0 = (InitValue(Np[0], 1.) + InitValue(Np[0], 15.) + 
      InitValue(Np[0], 31.) + InitValue(Np[0], 63.)) / 4.

# initialize the collection of all solution vectors on all grids
v = np.array([np.empty(Np[i]) for i in range(Ng)], 
             dtype=np.ndarray)

# set up coefficient matrix
A = np.array([Init1DLaplaceAb(Np[i])[0]/((2**i)**2) for i in range(Ng)], 
             dtype=np.ndarray)

# set up f vector (for f in Av=f)
f = np.array([Init1DLaplaceAb(Np[i])[1] for i in range(Ng)], 
             dtype=np.ndarray)

# set number of relaxing on each grids
n1 = 1; n2 = 1

# set up the array for saving transient solution
vT = np.empty((Ng-1)*n1+Ng*n2 ,dtype=np.ndarray)
```

The initial guess on the finest grid is:

```python
plt.plot(x[0], v0, label=r'$v_h$')
plotAssist(title='Initial Guess on the Finest Grid', ylmt=(-1., 1.))
```

Let's start to relax on multigrids now!

```python
# V-Cycle, upper half
c = 0
v[0] = v0.copy()
for i in range(Ng-1):
    for n in range(n1):
        v[i] = WJacobi(Np[i], A[i], v[i].copy(), f[i])
        vT[c] = v[i].copy()
        c += 1
    v[i+1] = F2C1D(Np[i], v[i])

# V-Cycle, lower half
for i in range(Ng-1, 0, -1):
    for n in range(n2):
        v[i] = WJacobi(Np[i], A[i], v[i].copy(), f[i])
        vT[c] = v[i].copy()
        c += 1
    v[i-1] = C2F1D(Np[i], v[i])

# relax on the finest grid
for n in range(n2):
    v[0] = WJacobi(Np[0], A[0], v[0].copy(), f[0])
    vT[c] = v[0].copy()
    c += 1
```

The result from weighted Jacobi method is also obtained for comparation.

```python
# initialize soln. vector
vWJ = v0.copy()

# initialize the array for intermediate soln.
vT_WJ = np.empty(c, dtype=np.ndarray)

# initialize the error array
errWJ = np.empty(c)

# start iterations
for i in range(c):
    vWJ = WJacobi(Np[0], A[0], vWJ.copy(), f[0])
    vT_WJ[i] = vWJ.copy()
```

```python
# plot the aproximated soln. in intermediate iteration
plt.figure(figsize=(18, 12))

for i in range(c):
    plt.subplot(2, 3, i+1)
    plt.plot(np.linspace(0., 1., np.size(vT[i])), vT[i], 'k-', label='V-Cycle, Nested Iteration')
    plt.plot(x[0], vT_WJ[i], 'r--', label='Weighted Jacobi Method')
    plotAssist(title='Iteration #'+str(i+1), ylmt=(-1., 1.))
```

From the above example, we can see that the multigrid method really converges faster than what weighted Jacobi method does! And the oscillatory errors converges faster than smooth error does in the multigrid method. However, we can also find that the most smooth error can not easily be eliminated by no matter nested iteration scheme or weighted Jacobi method.

## Correction Scheme
The second stragety -- correction scheme -- relax the residual equation instead of original equation, i.e. $Av=f$ on coarse grids. If we define error $e=u-v$, where $u$ is the analytical solutions on the grid point, and define residual $r=f-Av$, then the residual equation is:

$$Au-Av=f-Av~=>~A(u-v)=f-Av$$
thus
$$Ae=r$$

The correction scheme relax $Ae=r$ on coarse to obtain the error on coarse grids and then correct the solution, $v$, on fine grid by that errors. The scheme like V-cycle scheme can still be applied on the correction scheme: 

* Relax on $A^hv^h=f^h$ $n_1$ times on $\Omega^{h}$ with initial guess $v^h$
* Compute $r^{2h}=MAP_{f->c}(f^h-A^hv^h)$
* Relax on $A^{2h}e^{2h}=r^{2h}$ $n_1$ times on $\Omega^{2h}$ with initial guess of $e^{2h}$ being 0
* Compute $r^{4h}=MAP_{f->c}(r^{2h}-A^{2h}e^{2h})$

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\vdots$

* Compute $r^{Mh}=MAP_{f->c}(r^{(M/2)h}-A^{(M/2)h}e^{(M/2)h})$
* Rleax on $A^{Mh}e^{Mh}=r^{Mh}$ $n_1$ times on $\Omega^{Mh}$ with initial guess of $e^{Mh}$ being 0
* Compute $e^{(M/2)h}=e^{(M/2)h}+MAP_{c->f}(e^{M.h})$
* Rleax on $A^{(M/2)h}e^{(M/2).}=r^{(M/2)h}$ $n_2$ times on a $\Omega^{(M/2)h}$

$~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\vdots$

* Compute $e^{2h}=e^{2h}+MAP_{c->f}(e^{4h})$
* Relax on $A^{2h}e^{2h}=r^{2h}$ $n_1$ times on $\Omega^{2h}$
* Compute $v^{h}=v^{h}+MAP_{c->f}(e^{2h})$
* Relax on $A^hv^h=f^h$ $n_2$ times on $\Omega^{h}$ to obtain a final approximation to the solution

One thing needs to be notice, specifically in this Laplace example, is the coefficient matrix $A$. In nested iteration, because the $f$ in the equations we solved on all grids are all zeros, so we were actually solving $v_{i-1}-2v_{i}+v_{i+1}=0$ in our code. However, since now our constant terms in residual equations are not zero anymore, and what we are solving is $\left(e_{i-1}-2e_{i}+e_{i+1}\right)/dx^2=r_i$, we have to make a small change in the line that call the function to compute $A$.

```python
# set up the number of grid points
Nph = 65; 

# set the number of grids
Ng = 3

# set up the number of grid points for all grids
Np = np.array([(Nph-1)/(2**i)+1 for i in range(Ng)])

# set up coordinates
x = np.array([np.linspace(0., 1., Np[i]) for i in range(Ng)], 
             dtype=np.ndarray)

# set up initial guess for the finest grid
# v0 = 2 * np.random.random(Nph) - 1
v0 = (InitValue(Np[0], 1.) + InitValue(Np[0], 15.) + 
      InitValue(Np[0], 31.) + InitValue(Np[0], 63.)) / 4.

# initialize the collection of all solution vectors on all grids
v = np.array([np.empty(Np[i]) for i in range(Ng)], 
             dtype=np.ndarray)

# set up coefficient matrix
A = np.array([Init1DLaplaceAb(Np[i])[0]/((1./(Np[i]-1))**2) for i in range(Ng)], 
             dtype=np.ndarray)

# set up f vector (for f in Av=f)
f = np.array([Init1DLaplaceAb(Np[i])[1] for i in range(Ng)], 
             dtype=np.ndarray)

# set number of relaxing on each grids
n1 = 1; n2 = 1

# set up the array for saving transient solution
vT = np.empty(Ng*n1+(Ng-1)*n2 ,dtype=np.ndarray)
```

For our convenience in programming, we still use variables $v$ and $f$ when we relax the residual equations on coarse grids.

```python
# V-Cycle, relax Av=f on finest grid
c = 0
v[0] = v0.copy()

# V-Cycle, relax residual equation on coarse grids
# note, the 1st will relax Au=f on finest grid, rather than residual eq.
for i in range(Ng):
    for n in range(n1):
        v[i] = WJacobi(Np[i], A[i], v[i].copy(), f[i])
        vT[c] = v[i].copy()
        c += 1
    if (i < Ng-1):
        f[i+1] = F2C1D(Np[i], f[i]-np.dot(A[i], v[i]))
        v[i+1][:] = 0.
        
# V-Cycle, relax Av=f on coarse grid and back to finest one
for i in range(Ng-2, -1, -1):
    v[i] = v[i] + C2F1D(Np[i+1], v[i+1])
    for n in range(n2):
        v[i] = WJacobi(Np[i], A[i], v[i].copy(), f[i])
        vT[c] = v[i].copy()
        c += 1
```

Again, weighted Jacobi method is used for comparison.

```python
# initialize soln. vector
vWJ = v0.copy()

# initialize the array for intermediate soln.
vT_WJ = np.empty(c, dtype=np.ndarray)

# initialize the error array
errWJ = np.empty(c)

# start iterations
for i in range(c):
    vWJ = WJacobi(Np[0], A[0], vWJ.copy(), f[0])
    vT_WJ[i] = vWJ.copy() # err = 0 - v for Laplace PDE
```

To plot $v$ at different iteration, we have to do some post-processing. The data in $vT$ array are the solutions to residual equations on respective coarse grids, except the dada on $\Omega^h$. So manually post-processing is needed.

```python
vT[1] = F2C1D(Np[0], vT[0]) + vT[1] # after relax Ae=r on O2h
vT[2] = F2C1D(Np[1], vT[1]) + vT[2] # after relax Ae=r on h
vT[3] = F2C1D(Np[0], vT[0]) + vT[3] # after relax Av=f on O2h
```

```python
# plot the aproximated soln. in intermediate iteration
plt.figure(figsize=(18, 12))

for i in range(c):
    plt.subplot(2, 3, i+1)
    plt.plot(np.linspace(0., 1., np.size(vT[i])), vT[i], 'k-', label='V-Cycle, Nested Iteration')
    plt.plot(x[0], vT_WJ[i], 'r--', label='Weighted Jacobi Method')
    plotAssist(title='Iteration #'+str(i+1), ylmt=(-1., 1.))
```

We can see that the smooth error is still difficult to be eliminated. This means using 3 grids are not enough. Probably we need to use more coarse grids in order to eliminate the most smooth error..
